# Predicciones de test

```python
from sklearn.preprocessing import StandardScaler
```

```python
import pandas as pd
import pickle

with open("modelo_1.pkl", "rb") as f:
    lr_st = pickle.load(f)
```

```python
df_test = pd.read_csv("data/test.csv")
```

```python
df_test["Ram"] = df_test["Ram"].str.replace("GB", "").astype(float)
df_test["Weight"] = df_test["Weight"].str.replace("kg", "").astype(float)
```

```python
df_test.info()
```

```python
X = df_test[["Ram", "Weight", "Inches"]]
```

```python
st = StandardScaler()
st.fit(X)
```

```python
pred_st = st.transform([[16, 3.2, 16]])
```

```python
lr_st.predict(pred_st)
```

```python
X_st = st.transform(X)
```

```python
pred = lr_st.predict(X_st)
pred
```
